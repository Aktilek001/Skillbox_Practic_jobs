```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bubble.csv')
df = pd.DataFrame(df)

df['size2'] = df['size'] ** 2
```

```python
plt.plot(df['size'], df[' iterations'], label='Итерации')
plt.plot(df['size'], df['size2'], label='size^2')
plt.xlabel('Размер массива')
plt.ylabel('Количество операций')
plt.legend()
plt.show()
```

В худшем случае алгоритм делает n проходов по массиву.
В каждом проходе количество итераций уменьшается на 1.
Общее количество итераций:
iterations = n + (n - 1) + ... + 1 = n(n + 1) / 2
