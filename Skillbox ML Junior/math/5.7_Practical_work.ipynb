```python
# тут задача 1.3

from scipy.stats import norm

# Заданные значения
mu = 5
sigma = 3
a = 7
b = 12

# Вычисление вероятности попадания в интервал (7;12)
prob = norm.cdf(b, loc=mu, scale=sigma) - norm.cdf(a, loc=mu, scale=sigma)

print("Вероятность попадания случайной величины в интервал (7;12):", prob)
```

```python
# тут задача 1.4

import numpy as np
from scipy.integrate import quad

# Заданные значения
mu = 4
sigma_squared = 12

# Функция плотности вероятности
def f(x):
    return (1 / np.sqrt(2 * np.pi * sigma_squared)) * np.exp(-(x - mu)**2 / (2 * sigma_squared))

# Вычисление интеграла
integral_value, _ = quad(f, 6, 10)
print("Вероятность попадания случайной величины в интервал (6;10):", integral_value)
```

```python
# тут задача 2.1

import random

def generate_random_sequence(length, seed=None):
    random_sequence = []
    
    # Устанавливаем начальное значение для генерации
    if seed is not None:
        random.seed(seed)
    x = random.random()  # Произвольное число от 0 до 1
    
    # Генерация последовательности
    for _ in range(length):
        x = 4 * x * (1 - x)  # Мультипликативный метод
        random_sequence.append(x)
    
    return random_sequence

# Генерация последовательности из 100 случайных чисел
sequence = generate_random_sequence(100)

# Вывод сгенерированной последовательности
print(sequence)
```

```python
# тут задача 2.2

import math

def generate_exponential_random_numbers(n):
    random_numbers = []
    for _ in range(n):
        u = random.random()  # Генерируем случайное число из равномерного распределения [0, 1]
        y = - (1/4) * math.log(1 - u)  # Применяем обратную функцию распределения
        random_numbers.append(y)
    return random_numbers

# Генерируем 10 случайных чисел
random_numbers = generate_exponential_random_numbers(10)

# Выводим результат
print(random_numbers)
```
