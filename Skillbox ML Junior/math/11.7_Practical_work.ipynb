***Задание 1***

```python
from sympy import Matrix

# Уравнения из задачи
equations_a = [(Matrix([[2, 3], [0, 1]]), Matrix([2, 4])),
               (Matrix([[1, 5], [0, 3]]), Matrix([3, 9]))]

equations_b = [(Matrix([[2, 3, -1], [0, 1, 1], [1, -1, 2]]), Matrix([2, 4, 6])),
               (Matrix([[1, 5, -2], [0, 3, 1], [-4, -6, 2]]), Matrix([8, 16, 32]))]

# Решение уравнений
solutions_a = []
solutions_b = []

# Решение для задачи а)
for A, B in equations_a:
    X = A.inv() * B
    solutions_a.append(X)

# Решение для задачи б)
for A, B in equations_b:
    X = A.inv() * B
    solutions_b.append(X)

# Вывод решений
print("Решения для задачи а):")
for i, solution in enumerate(solutions_a, 1):
    print(f"Уравнение {i}: {solution}")

print("\nРешения для задачи б):")
for i, solution in enumerate(solutions_b, 1):
    print(f"Уравнение {i}: {solution}")
```

***Задание 2***

```python
# Уравнение 1
A1 = Matrix([[2, 5], [7, -3]])
B1 = Matrix([11, -23])
X1 = A1.pinv() * B1

# Уравнение 2
A2 = Matrix([[-3, 1], [3, 5]])
B2 = Matrix([-2, 8])
X2 = A2.pinv() * B2

# Уравнение 3
A3 = Matrix([[2, 3], [3, -1]])
B3 = Matrix([12, 7])
X3 = A3.pinv() * B3

# Уравнение 4
A4 = Matrix([[1, 1, 2], [2, -1, -2], [4, 1, 4]])
B4 = Matrix([-1, -4, -2])
X4 = A4.pinv() * B4

print("Решение для уравнения 1:", X1)
print("Решение для уравнения 2:", X2)
print("Решение для уравнения 3:", X3)
print("Решение для уравнения 4:", X4)
```

```python
# Ожидаемые значения решений в виде матриц
expected_values = [Matrix([[2], [1]]), Matrix([[-1], [3]]), Matrix([[3], [5]]), Matrix([[1], [-2], [1]])]

# Полученные значения решений
X_values = [X1, X2, X3, X4]

# Вычисляем среднеквадратичную ошибку для каждой системы
for i, X in enumerate(X_values, 1):
    expected = expected_values[i-1]
    mse = (X - expected).norm() ** 2
    print(f"Среднеквадратичная ошибка для уравнения {i}: {mse}")
```

```python
from scipy.stats import mannwhitneyu

# Заданные выборки X и Y
X = [-2, 1, 4, -1, 5, 7, 9, 0, 12]
Y = [6, -3, 2, 1, 0, 4, 2]

# Выполнение теста Манна-Уитни
stat, p_value = mannwhitneyu(X, Y)

# Вывод результатов
print("Статистика теста:", stat)
print("p-значение:", p_value)

# Определение статистической значимости
alpha = 0.05
if p_value < alpha:
    print("Отвергаем нулевую гипотезу: существует статистически значимое различие между распределениями X и Y.")
else:
    print("Не удалось отвергнуть нулевую гипотезу: нет статистически значимых различий между распределениями X и Y.")
```
